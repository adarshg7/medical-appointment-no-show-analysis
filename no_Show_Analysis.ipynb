```python
import pandas as pd
import statsmodels.api as sm

# 1. Prepare Features (X) and Target (y)
# Selecting the key features we've analyzed
features = ['Age', 'Scholarship', 'SMS_received', 'Wait_Days']
X = df[features]
y = df['No-show']

# 2. Add a Constant (Interupt) - Required by statsmodels
X = sm.add_constant(X)

# 3. Fit the Logistic Regression Model
log_reg = sm.Logit(y, X).fit()

# 4. Show the Summary
print(log_reg.summary())

# 5. Calculate Odds Ratios
import numpy as np
params = log_reg.params
conf = log_reg.conf_int()
conf['Odds Ratio'] = params
conf.columns = ['2.5%', '97.5%', 'Odds Ratio']
print(np.exp(conf))
```
